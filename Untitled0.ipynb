```
train_img_folder='train_img/train_images'
train_labs_folder='train_labs/train_segmentations'
# print(os.listdir('train_img'))
# print(os.listdir('train_labs'))

batch_size = 32
img_height = 256
img_width = 256

```

```
def data_gen(img_folder, mask_folder, batch_size):
  c = 0
  n = sorted(os.listdir(img_folder)) #List of training images
  # n2 = sorted(os.listdir(mask_folder)) #List of training images
  random.shuffle(n)
  
  while (True):
    img = np.zeros((batch_size, img_height, img_width, 3)).astype('float')
    mask = np.zeros((batch_size, img_height, img_width, 20)).astype('float')

    for i in range(c, c+batch_size): #initially from 0 to 16, c = 0. 
      # print(n[i],n2[i])
      train_img = cv2.imread(img_folder+'/'+n[i][:-4]+'.jpg')
      train_img = cv2.cvtColor(train_img, cv2.COLOR_BGR2RGB)
      train_img =  cv2.resize(train_img, (img_height, img_width))# Read an image from folder and resize
      
      img[i-c] = train_img #add to array - img[0], img[1], and so on.
                                                   
      
      train_mask = cv2.imread(mask_folder+'/'+n[i][:-4]+'.png')
      train_mask = cv2.cvtColor(train_mask, cv2.COLOR_BGR2RGB)
      train_mask_seg=np.zeros([train_mask.shape[0],train_mask.shape[1],20],dtype=np.uint8)
   
      for j in range(20):
        a=np.zeros([20],dtype=np.uint8)
        a[j]=1
        train_mask_seg[train_mask[:,:,0]==j]=a
      train_mask=train_mask_seg
      train_mask = cv2.resize(train_mask, (img_height, img_width))
      # train_mask = train_mask.reshape(img_height, img_width, 1) # Add extra dimension for parity with train_img size [512 * 512 * 3]

      mask[i-c] = train_mask

    c+=batch_size
    if(c+batch_size>=len(os.listdir(img_folder))):
      c=0
      random.shuffle(n)
                  # print "randomizing again"
    yield img, mask
```

creating a keras model to classify the data

```
from keras.applications.resnet50 import ResNet50

resnet = ResNet50(include_top=False, weights='imagenet', input_shape=(img_height,img_width,3))
# resnet.summary()
out_resnet=resnet.get_layer('conv4_block5_out').output

down_stack =keras.Model(inputs=resnet.input,outputs=out_resnet)
down_stack.trainable = False
def create_model(out_channels=20):
  inputs =keras.layers.Input(shape=[None,None,3])
  x=inputs 
  x=down_stack(x)

  


  x=keras.layers.Conv2D(512,[5,5],padding='same',dilation_rate=2,activation='relu')(x)
  x=keras.layers.BatchNormalization()(x)
  x=keras.layers.Conv2D(256,[5,5],padding='same',dilation_rate=2,activation='relu')(x)
  x=keras.layers.BatchNormalization()(x)
  x=keras.layers.Conv2DTranspose(128,kernel_size=[4,4],strides=[4,4],activation='relu')(x)
  x=keras.layers.BatchNormalization()(x)
  x=keras.layers.Conv2DTranspose(64,kernel_size=[2,2],strides=[2,2],activation='relu')(x)
  x=keras.layers.BatchNormalization()(x)
  x=keras.layers.Conv2DTranspose(32,kernel_size=[2,2],strides=[2,2],activation='relu')(x)
  x=keras.layers.BatchNormalization()(x)
  x=keras.layers.Conv2D(20,[3,3],padding='same',dilation_rate=1,activation='relu')(x)
  x=keras.layers.Conv2D(20,[1,1],padding='same',dilation_rate=1,activation='softmax')(x)
  model=keras.Model(inputs=inputs, outputs=x)

  return model


model =create_model(20)

```
